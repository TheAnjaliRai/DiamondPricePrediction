```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Gender']])
gender_df = pd.DataFrame(encoded.toarray(),columns = encoder.get_feature_names_out())
```

```python
df = pd.concat([df,gender_df],axis = 1)
```

```python
dict = {'Normal Weight' : 1 , 'Underweight' : 0 , 'Overweight' : 2 ,'Obese': 3}
```

```python
df['Label_encoded'] = df['Label'].map(dict)
```

```python
df = df.drop("Gender" , axis = 1)
df = df.drop('Label' , axis = 1)
```

```python
X = df.iloc[:,:-1]
y = df['Label_encoded']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y, test_size = 0.20,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs')
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score,classification_report
print(confusion_matrix(y_pred , y_test ))
print(accuracy_score(y_pred , y_test ))
print(classification_report(y_pred , y_test ))
```
